```python
#This code is going to open the normalized intensity files and is going to extract the intensity values
#at the end of the trace and the maximum including the position at which the maximum is found. Then,
#it calculates de difference of intesities from the max and the end intensities, this strategy is used because 
#in some experiments specially in the cases where the decay is small, the difference in intensity from the point 
#where the NETs form and the decay caused by the expansion of the NET, it fits to the exponential decay understimating
#the halflife because what it is really quantifying is the expansion of the NET. To overcome this we are taking the 
#Delta Intensity Max and Intensity End.

#I have also cataloged the cells based on the position where the max intensity is and Delta intensity


import xlrd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import csv
import math

directory ='/Users/arambui/Desktop/To analyse/20201124_timelapse_plasma_screen/plate1_quantification/with PMA'
#This for look will analyse the CSV files named Results_filename.csv present in the directory 
#this files are the output from the quantification in fiji


for filename in os.listdir(directory):
    if filename.endswith(".csv") and filename.startswith("NormMin_"):
        print(filename)
        csv_file=filename
        newname=csv_file[-22:-12]
        print(newname)
        #pd.read_csv(csv_file, sep=":").T
        df = pd.read_csv(csv_file, header=None).T #header None helps to get all the columns avoiding the first one to be used as the index
        #print(df.head(5))
        #I make a list to accumulate the coefficients obtained from the different columns (Iend,Imax,DeltaIntensity)
        Iend_acc=[]
        Imax_acc=[]
        DeltaInt_acc=[]
        MaxValuePos=[]
        N=int(len(df))
        print(N)
        Nlate=int(0.75*N)
        print(Nlate)
        
        # For look were it looks at each raw of a column and it identifies the end point and the max point
        for c in df.columns:
            Iend=df[c].iloc[-1]
            #print(Iend)
            max_point = max(df[c])
            #print(max_point)
            DeltaInt = max_point - Iend
            #maxValuIndex will provide the postion in time where the max value is found , when NET is formed.
            maxValueIndex = df[c].idxmax() 
            #print(DeltaInt)
            #accumulate the values obtained from each column
            Iend_acc.append(Iend)
            Imax_acc.append(max_point)
            DeltaInt_acc.append(DeltaInt)
            MaxValuePos.append(maxValueIndex)
            #Make a new dataframe with the accumulated values
        df_Iend_acc=pd.DataFrame(Iend_acc)
        df_Imax_acc=pd.DataFrame(Imax_acc)
        #print(df_Imax_acc)
        df_DeltaInt_acc=pd.DataFrame(DeltaInt_acc)
        df_MaxValuePos=pd.DataFrame(MaxValuePos)
        #print(df_DeltaInt_acc)
        #print(df_MaxValuePos)
        df_coef_out = pd.concat([df_Iend_acc, df_Imax_acc,df_MaxValuePos, df_DeltaInt_acc], axis=1, join_axes=[df_Iend_acc.index])
        df_coef_out.columns=["Iend", "Imax","ImaxPos", "DeltaInt"]
        #print(df_coef_out)
        df_coef_out.loc[df_coef_out['DeltaInt'] > 0.05, 'Necrotic/NETotic/Late Netosis'] = 'Netotic cell'
        df_coef_out.loc[(df_coef_out['ImaxPos'] >= N-6)|(df_coef_out['DeltaInt'] < 0.05), 'Necrotic/NETotic/Late Netosis'] = 'dead cell'
        df_coef_out.loc[(df_coef_out['ImaxPos'] >= Nlate)&(df_coef_out['ImaxPos']< N-6), 'Necrotic/NETotic/Late Netosis'] = 'Late_NETotic cell'
        print(df_coef_out)
        df_coef_out.to_csv('All_cells_'+newname+".csv", index=False)
        # Create variable with TRUE if NETotic cells
        df_NETotic=(df_coef_out.loc[df_coef_out['Necrotic/NETotic/Late Netosis'] == 'Netotic cell'])
        df_NETotic.to_csv('NETotic_'+newname+".csv", index=False)
 
    
```
